```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imshow, imread

img = cv2.imread("Cat.jpg")
imshow(img)
```

```python
img = cv2.imread("Cat.jpg",0)
imshow(img)
```

```python
bright_increase = np.clip(img + 50, 0, 255).astype(np.uint8)
plt.imshow(bright_increase, cmap='gray')
plt.title("Brightness Improved")
plt.axis("off")
plt.show()
```

```python
bright_decrease = np.clip(img - 50, 0, 255).astype(np.uint8)
plt.imshow(bright_decrease, cmap='gray')
plt.title("Brightness Reduced")
plt.axis("off")
plt.show()
```

```python
_, thresh = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
plt.imshow(thresh, cmap='gray')
plt.title("Thresholded Image")
plt.axis("off")
plt.show()
```

```python
neg = 255 - img
plt.imshow(neg, cmap='gray')
plt.title("Negative Image")
plt.axis("off")
plt.show()
```

```python
c = 255 / np.log(1 + np.max(img))
log_transformed = c * np.log(1 + img)
log_transformed = np.array(log_transformed, dtype=np.uint8)
plt.imshow(log_transformed, cmap='gray')
plt.title("Log Transformation")
plt.axis("off")
plt.show()
```

```python
gamma = 0.5 
gamma_corrected = np.array(255 * (img / 255) ** gamma, dtype=np.uint8)
plt.imshow(gamma_corrected, cmap='gray')
plt.title("Gamma Transformation")
plt.axis("off")
plt.show()
```
