```python
import numpy as np
import matplotlib.pyplot as plt 
from skimage import img_as_uint
from skimage.io import imshow, imread
from skimage.color import rgb2hsv
from skimage.color import rgb2gray

array_1 = np.array([[255, 0],[0, 255]])
imshow(array_1, cmap ='gray');
```

```python
array_2 = np.array([[255, 0, 255],[0, 255, 0],[255, 0, 255]])
imshow(array_2, cmap = 'grey');
```

```python
array_spectrum = np.array([np.arange(0,255,17),np.arange(255,0,-17),np.arange(0,255,17),np.arange(255,0,-17)])
fig,ax =plt.subplots(1, 2,figsize=(12,4))
ax[0].imshow(array_spectrum, cmap= 'gray')
ax[0].set_title('ArangeGeneration')
ax[1].imshow(array_spectrum.T, cmap= 'gray')
ax[1].set_title('Transpose Generation');
```

```python
array_colors = np.array([[[255, 0, 0],[0, 255, 0],[0, 0, 255]]])
imshow(array_colors);
```

```python
fig, ax = plt.subplots(1, 3,figsize=(6,4),sharey= False)

ax[0].imshow(nature3[:, 0:99])
ax[0].set_title('First Split')

ax[1].imshow(nature3[:, 99:198])
ax[1].set_title('Second Split')

ax[2].imshow(nature3[:, 198:297])
ax[2].set_title('Third Split');
```

```python
imshow(nature3[10:140, 50:200]);
```

```python
fig, ax = plt.subplots(1, 3, figsize=(12,4), sharey =True)
ax[0].imshow(nature3[:,:,0], cmap='Reds')
ax[0].set_title('Red')

ax[1].imshow(nature3[:,:,1],cmap='Greens')
ax[1].set_title('Green')

ax[2].imshow(nature3[:,:,2],cmap='Blues')
ax[2].set_title('Blue');
```

```python
nature3_hsv = rgb2hsv(nature3)
fig, ax =plt.subplots(1, 3, figsize=(12,4), sharey = True)

ax[0].imshow(nature3_hsv[:,:,0], cmap='hsv')
ax[0].set_title('Hue')

ax[1].imshow(nature3_hsv[:,:,1],cmap='gray')
ax[1].set_title('Saturation')

ax[2].imshow(nature3_hsv[:,:,2], cmap='gray')
ax[2].set_title('Value');
```

```python
nature3_gray = rgb2gray(nature3)
fig, ax =plt.subplots(1, 5, figsize=(17,6), sharey = True)

ax[0].imshow(nature3_gray, cmap = 'gray')
ax[0].set_title('GrayscaleOriginal')

ax[1].imshow(img_as_uint(nature3_gray >0.25),cmap = 'gray')
ax[1].set_title('Greater than0.25')

ax[2].imshow(img_as_uint(nature3_gray >0.50),cmap = 'gray')
ax[2].set_title('Greater than0.50');

ax[3].imshow(img_as_uint(nature3_gray >0.75),cmap = 'gray')
ax[3].set_title('Greater than0.75')

ax[4].imshow(img_as_uint(nature3_gray >np.mean(nature3_gray)),cmap = 'gray')
ax[4].set_title('Greater than Mean');
```
